```python
# 연령대 구간 (bins) 설정
def categorize_age(age):
    if age < 20:
        return '10대'
    elif age < 30:
        return '20대'
    elif age < 40:
        return '30대'
    elif age < 50:
        return '40대'
    elif age < 60:
        return '50대'
    elif age < 60:
        return '60대 이상'
    else:
        return '70대 이상'

df['연령대'] = df['나이'].apply(categorize_age)
```

```python
# 계절 구분 (봄: 3~5월, 여름: 6~8월, 가을: 9~11월, 겨울: 12~2월)
def get_season(date):
    month = int(date.split("-")[1])
    if month in [3, 4, 5]:
        return "봄"
    elif month in [6, 7, 8]:
        return "여름"
    elif month in [9, 10, 11]:
        return "가을"
    else:
        return "겨울"

df["구매 시즌"] = df["구매 날짜"].apply(get_season)

print(df[["구매 날짜", "구매 시즌"]].head())
```

```python
# 고객 등급을 점수로 변환 (VIP: 3, 일반: 2, 신규: 1, 이탈가능: 0)
grade_mapping = {"vip": 3, "일반": 2, "신규": 1, "이탈가능": 0}
df["등급 점수"] = df["고객세그먼트"].map(grade_mapping)
```

```python
# 🔹 금액 컬럼을 숫자로 변환
df['누적 구매 금액'] = pd.to_numeric(df['누적 구매 금액'], errors='coerce').fillna(0)

# 🔹 누적 구매 금액에 따른 점수 부여 (1-5점)
def get_amount_score(amount):
    if amount < 30000000:
        return 1
    elif amount < 60000000:
        return 2
    elif amount < 90000000:
        return 3
    elif amount < 120000000:
        return 4
    else:
        return 5

# 🔹 기존 등급에 따른 점수 부여 (0-3점)
grade_mapping = {"vip": 3, "일반": 2, "신규": 1, "이탈가능": 0}

# 🔹 최종 고객 등급 점수 계산 함수 (1-10등급)
def calculate_final_grade(amount_score, grade_score):
    total_score = amount_score + grade_score
    return np.clip(total_score, 1, 10)

# 🔹 누적 구매 금액에 따른 점수 계산
df['금액 점수'] = df['누적 구매 금액'].apply(get_amount_score)


# 🔹 최종 고객 등급 점수 계산 (1-10등급)
df['고객 등급 점수'] = df.apply(lambda row: calculate_final_grade(row['금액 점수'], row['등급 점수']), axis=1)

# 🔹 결과 확인
print("\n📌 [고객 등급 점수 계산 결과]")
print(df[['누적 구매 금액', '고객세그먼트', '금액 점수', '등급 점수', '고객 등급 점수']].head())
```

# 국내 신차 구매 시 할부 여부 기준 

---

## 1️. 4천만 원 이상 차량은 대부분 할부/리스 이용
 **신차 구매 가격이 4천만 원을 넘으면 할부 이용률 증가**  
 **대부분 중형차급 이상 (SUV, 세단, 전기차)에서 적용됨**

|  **차종**             |  **가격대 (예시)** |  **할부 이용 여부**         |
|------------------------|-----------------|----------------------|
| 현대 아반떼 (CN7 N)   | **약 3,500만 원** |  할부 이용 가능 (40%)  |
| 기아 쏘렌토 (HEV)     | **약 4,500만 원** |  할부 이용 확률 높음 (70%) |
| 제네시스 G80         | **약 6,500만 원 이상** |  대부분 할부/리스 이용 |

---

>  **할부 기준 공식**
> $$LTV = \frac{총 할부 금액}{구매 가격} \times 100\%$$
>
> 🔍 **할부율이 높은 이유?**
> - 초기 비용 부담 완화  
> - 저금리 자동차 금융 상품 활성화  
> - 월 할부 상환 가능 고객 증가  

---

## 2️. 자동차 구매 시 할부 금융 이용률 ( 금융감독원 기준)
 **국내 자동차 구매 시 할부 금융 이용률 (금융통계정보 시스템 기준)**  
 **국산차 평균 할부율:**  
- **3천만 원 이하 차량** →  약 **40%가 할부 사용**  
- **4천만 원 이상 차량** →  **70% 이상이 할부 사용**  
- **고가 차량일수록 현금 일시불 구매 비율이 낮음**  


---

## 3️.  결론 
 **4천만 원 이상 차량의 경우 할부 가능성이 높음**  
 **국산차 구매 시 4천만 원이 기준선이 될 수 있음**  
 **데이터 기반 분석을 통해 차량 가격과 금융 패턴을 예측 가능**

```python
# 🔹 1. 구매 횟수에 따른 점수 부여 (1~5점)
def get_purchase_count_score(count):
    if count >= 5:
        return 5
    elif count >= 4:
        return 4
    elif count >= 3:
        return 3
    elif count >= 2:
        return 2
    else:
        return 1

# 🔹 2. 누적 구매 금액에 따른 점수 부여 (1~5점)
def get_amount_score(amount):
    if amount >= 500000000:  # 5억 이상
        return 5
    elif amount >= 300000000:  # 3억 이상
        return 4
    elif amount >= 150000000:  # 1억 5천만 이상
        return 3
    elif amount >= 50000000:  # 5천만 이상
        return 2
    else:
        return 1  # 5천만 미만


# 🔹 3. 구매 빈도 점수 (최근 구매 고객 가산점, 1~5점)
def get_recency_score(days_since_last_purchase):
    if days_since_last_purchase <= 30:
        return 5  # 한 달 이내
    elif days_since_last_purchase <= 90:
        return 4  # 3개월 이내
    elif days_since_last_purchase <= 180:
        return 3  # 6개월 이내
    elif days_since_last_purchase <= 365:
        return 2  # 1년 이내
    else:
        return 1  # 1년 이상 미구매


# 🔹 4. 거래 방식 점수 (할부보다는 일시불 선호)
def get_transaction_score(method):
    return 2 if method == "일시불" else 1

# 🔹 5. 구매 경로 점수 (온라인 우선)
def get_purchase_path_score(path):
    return 2 if path == "온라인" else 1

# 🔹 6. 기존 등급 점수 활용 (이미 계산된 고객 등급 점수)
def get_customer_grade_score(grade_score):
    return grade_score  # 고객 등급 점수를 그대로 사용

# 🔹 데이터 불러오기 (기존 DataFrame 사용)
df["금액 점수"] = df["누적 구매 금액"].apply(get_amount_score)
df["구매 횟수 점수"] = df["구매 횟수"].apply(get_purchase_count_score)

# 🔹 최근 구매일 계산 (현재 날짜 기준, 예: 2025-03-20)
current_date = pd.Timestamp("2025-03-20")
df["구매 날짜"] = pd.to_datetime(df["구매 날짜"], errors="coerce")
df["마지막 구매 이후 일수"] = (current_date - df["구매 날짜"]).dt.days.fillna(9999)

df["구매 빈도 점수"] = df["마지막 구매 이후 일수"].apply(get_recency_score)
df["거래 방식 점수"] = df["거래 방식"].apply(get_transaction_score)
df["구매 경로 점수"] = df["구매 경로"].apply(get_purchase_path_score)

# 🔹 고객 충성도 지수 계산 (가중 평균 적용)
df["고객 충성도 지수"] = (
    df["금액 점수"] * 0.25 +  # 구매 금액 가중치 25%
    df["구매 횟수 점수"] * 0.2 +  # 구매 횟수 가중치 20%
    df["구매 빈도 점수"] * 0.2 +  # 최근 구매일 가중치 20%
    df["거래 방식 점수"] * 0.15 +  # 거래 방식 가중치 15%
    df["구매 경로 점수"] * 0.1 +  # 구매 경로 가중치 10%
    df["고객 등급 점수"] * 0.1  # 기존 고객 등급 점수 활용
)

# 🔹 최종 점수를 1~100 스케일로 변환 (Min-Max 정규화)
df["고객 충성도 지수"] = df["고객 충성도 지수"] / df["고객 충성도 지수"].max() * 100

# 🔹 결과 확인
print("\n📌 [고객 충성도 지수 계산 결과]")
print(df[["고객ID", "금액 점수", "구매 횟수 점수", "구매 빈도 점수", "거래 방식 점수", "구매 경로 점수", "고객 등급 점수", "고객 충성도 지수"]].head())



```

```python
#  현재 날짜 기준 설정 (예: 2025-03-20)
current_date = pd.Timestamp("2025-03-20")

#  날짜 데이터 변환
df["가입일"] = pd.to_datetime(df["가입일"], errors="coerce")
df["구매 날짜"] = pd.to_datetime(df["구매 날짜"], errors="coerce")

#  마지막 구매 이후 일수 계산
df["마지막 구매 이후 일수"] = (current_date - df["구매 날짜"]).dt.days.fillna(9999)

#  등급 분류 함수 (기준 업데이트됨)
def classify_customer(row):
    # 1️⃣ 신규 고객: 구매횟수 1회 & 가입일과 구매일이 최근 3개월 이내
    if row["구매 횟수"] == 1 and (current_date - row["구매 날짜"]).days <= 90:
        return "신규"

    # 2️⃣ VIP 고객: 고객 충성도 지수 80 이상 & 구매 횟수 5회 이상 & 1년 이내 구매
    elif row["고객 충성도 지수"] >= 80 and row["구매 횟수"] >= 4 and row["마지막 구매 이후 일수"] <= 365:
        return "vip"

    # 3️⃣ 일반 고객: 고객 충성도 지수 50~79 & 구매 횟수 3회 이상 & 1년 이내 구매
    elif 50 <= row["고객 충성도 지수"] < 80 and row["구매 횟수"] >= 3 and row["마지막 구매 이후 일수"] <= 365:
        return "일반"

    # 4️⃣ 이탈 가능 고객: 고객 충성도 지수 40 이하 또는 2년 이상 미구매
    elif row["고객 충성도 지수"] <= 40 or row["마지막 구매 이후 일수"] > 730:
        return "이탈가능"

    # 기본값 (일반 고객으로 분류)
    return "일반"

#  등급 컬럼 생성
df["등급"] = df.apply(classify_customer, axis=1)

#  등급 분포 확인
print("\n📌 [고객 등급 분포]")
print(df["등급"].value_counts())

#  결과 확인 (샘플 데이터 출력)
print("\n📌 [등급 할당 후 샘플 데이터]")
print(df[["고객ID", "고객 충성도 지수", "구매 횟수", "가입일", "구매 날짜", "마지막 구매 이후 일수", "등급"]].head(10))
```

# 📌 고객 생애 가치(CLV, Customer Lifetime Value)란?
고객 생애 가치(CLV)는 한 명의 고객이 기업과 거래하는 동안 발생할 총 수익을 예측하는 지표입니다.  
이 값이 높을수록 기업에 장기적으로 기여하는 고객이라는 의미입니다.  

---

## 📌 💡 CLV 계산 공식  

\[
\text{CLV} = \text{평균 구매 금액} \times \text{구매 빈도} \times \text{평균 고객 수명}
\]

---

## 📌 CLV 주요 개념  

### ✅ 평균 구매 금액 (Average Purchase Value)  
- 고객이 한 번 구매할 때 평균적으로 지출하는 금액  

### ✅ 구매 빈도 (Purchase Frequency)  
- 고객이 일정 기간 동안 얼마나 자주 구매하는지  

### ✅ 평균 고객 수명 (Customer Lifetime)  
- 고객이 기업과 평균적으로 거래하는 기간  

---

## 📌 CLV의 중요성  
- 고객의 **장기적 가치를 평가**하여 마케팅 전략을 최적화  
- 충성도가 높은 고객을 찾아 **리텐션 전략(재구매 유도)** 수립 가능  
- 기업의 **수익 예측 및 고객 세분화**에 활용
