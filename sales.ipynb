```python
# ğŸ”¹ ê¸ˆì•¡ ì»¬ëŸ¼ì„ ìˆ«ìë¡œ ë³€í™˜
df['ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡'] = pd.to_numeric(df['ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡'], errors='coerce').fillna(0)

# ğŸ”¹ ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡ì— ë”°ë¥¸ ì ìˆ˜ ë¶€ì—¬ (1-5ì )
def get_amount_score(amount):
    if amount < 30000000:
        return 1
    elif amount < 60000000:
        return 2
    elif amount < 90000000:
        return 3
    elif amount < 120000000:
        return 4
    else:
        return 5

# ğŸ”¹ ê¸°ì¡´ ë“±ê¸‰ì— ë”°ë¥¸ ì ìˆ˜ ë¶€ì—¬ (0-3ì )
grade_mapping = {"vip": 3, "ì¼ë°˜": 2, "ì‹ ê·œ": 1, "ì´íƒˆê°€ëŠ¥": 0}

# ğŸ”¹ ìµœì¢… ê³ ê° ë“±ê¸‰ ì ìˆ˜ ê³„ì‚° í•¨ìˆ˜ (1-10ë“±ê¸‰)
def calculate_final_grade(amount_score, grade_score):
    total_score = amount_score + grade_score
    return np.clip(total_score, 1, 10)

# ğŸ”¹ ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡ì— ë”°ë¥¸ ì ìˆ˜ ê³„ì‚°
df['ê¸ˆì•¡ ì ìˆ˜'] = df['ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡'].apply(get_amount_score)


# ğŸ”¹ ìµœì¢… ê³ ê° ë“±ê¸‰ ì ìˆ˜ ê³„ì‚° (1-10ë“±ê¸‰)
df['ê³ ê° ë“±ê¸‰ ì ìˆ˜'] = df.apply(lambda row: calculate_final_grade(row['ê¸ˆì•¡ ì ìˆ˜'], row['ë“±ê¸‰ ì ìˆ˜']), axis=1)

# ğŸ”¹ ê²°ê³¼ í™•ì¸
print("\nğŸ“Œ [ê³ ê° ë“±ê¸‰ ì ìˆ˜ ê³„ì‚° ê²°ê³¼]")
print(df[['ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡', 'ê³ ê°ì„¸ê·¸ë¨¼íŠ¸', 'ê¸ˆì•¡ ì ìˆ˜', 'ë“±ê¸‰ ì ìˆ˜', 'ê³ ê° ë“±ê¸‰ ì ìˆ˜']].head())
```

```python
# ğŸ”¹ 1. êµ¬ë§¤ íšŸìˆ˜ì— ë”°ë¥¸ ì ìˆ˜ ë¶€ì—¬ (1~5ì )
def get_purchase_count_score(count):
    if count >= 5:
        return 5
    elif count >= 4:
        return 4
    elif count >= 3:
        return 3
    elif count >= 2:
        return 2
    else:
        return 1

# ğŸ”¹ 2. ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡ì— ë”°ë¥¸ ì ìˆ˜ ë¶€ì—¬ (1~5ì )
def get_amount_score(amount):
    if amount >= 500000000:  # 5ì–µ ì´ìƒ
        return 5
    elif amount >= 300000000:  # 3ì–µ ì´ìƒ
        return 4
    elif amount >= 150000000:  # 1ì–µ 5ì²œë§Œ ì´ìƒ
        return 3
    elif amount >= 50000000:  # 5ì²œë§Œ ì´ìƒ
        return 2
    else:
        return 1  # 5ì²œë§Œ ë¯¸ë§Œ


# ğŸ”¹ 3. êµ¬ë§¤ ë¹ˆë„ ì ìˆ˜ (ìµœê·¼ êµ¬ë§¤ ê³ ê° ê°€ì‚°ì , 1~5ì )
def get_recency_score(days_since_last_purchase):
    if days_since_last_purchase <= 30:
        return 5  # í•œ ë‹¬ ì´ë‚´
    elif days_since_last_purchase <= 90:
        return 4  # 3ê°œì›” ì´ë‚´
    elif days_since_last_purchase <= 180:
        return 3  # 6ê°œì›” ì´ë‚´
    elif days_since_last_purchase <= 365:
        return 2  # 1ë…„ ì´ë‚´
    else:
        return 1  # 1ë…„ ì´ìƒ ë¯¸êµ¬ë§¤


# ğŸ”¹ 4. ê±°ë˜ ë°©ì‹ ì ìˆ˜ (í• ë¶€ë³´ë‹¤ëŠ” ì¼ì‹œë¶ˆ ì„ í˜¸)
def get_transaction_score(method):
    return 2 if method == "ì¼ì‹œë¶ˆ" else 1

# ğŸ”¹ 5. êµ¬ë§¤ ê²½ë¡œ ì ìˆ˜ (ì˜¨ë¼ì¸ ìš°ì„ )
def get_purchase_path_score(path):
    return 2 if path == "ì˜¨ë¼ì¸" else 1

# ğŸ”¹ 6. ê¸°ì¡´ ë“±ê¸‰ ì ìˆ˜ í™œìš© (ì´ë¯¸ ê³„ì‚°ëœ ê³ ê° ë“±ê¸‰ ì ìˆ˜)
def get_customer_grade_score(grade_score):
    return grade_score  # ê³ ê° ë“±ê¸‰ ì ìˆ˜ë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©

# ğŸ”¹ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° (ê¸°ì¡´ DataFrame ì‚¬ìš©)
df["ê¸ˆì•¡ ì ìˆ˜"] = df["ëˆ„ì  êµ¬ë§¤ ê¸ˆì•¡"].apply(get_amount_score)
df["êµ¬ë§¤ íšŸìˆ˜ ì ìˆ˜"] = df["êµ¬ë§¤ íšŸìˆ˜"].apply(get_purchase_count_score)

# ğŸ”¹ ìµœê·¼ êµ¬ë§¤ì¼ ê³„ì‚° (í˜„ì¬ ë‚ ì§œ ê¸°ì¤€, ì˜ˆ: 2025-03-20)
current_date = pd.Timestamp("2025-03-20")
df["êµ¬ë§¤ ë‚ ì§œ"] = pd.to_datetime(df["êµ¬ë§¤ ë‚ ì§œ"], errors="coerce")
df["ë§ˆì§€ë§‰ êµ¬ë§¤ ì´í›„ ì¼ìˆ˜"] = (current_date - df["êµ¬ë§¤ ë‚ ì§œ"]).dt.days.fillna(9999)

df["êµ¬ë§¤ ë¹ˆë„ ì ìˆ˜"] = df["ë§ˆì§€ë§‰ êµ¬ë§¤ ì´í›„ ì¼ìˆ˜"].apply(get_recency_score)
df["ê±°ë˜ ë°©ì‹ ì ìˆ˜"] = df["ê±°ë˜ ë°©ì‹"].apply(get_transaction_score)
df["êµ¬ë§¤ ê²½ë¡œ ì ìˆ˜"] = df["êµ¬ë§¤ ê²½ë¡œ"].apply(get_purchase_path_score)

# ğŸ”¹ ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜ ê³„ì‚° (ê°€ì¤‘ í‰ê·  ì ìš©)
df["ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜"] = (
    df["ê¸ˆì•¡ ì ìˆ˜"] * 0.25 +  # êµ¬ë§¤ ê¸ˆì•¡ ê°€ì¤‘ì¹˜ 25%
    df["êµ¬ë§¤ íšŸìˆ˜ ì ìˆ˜"] * 0.2 +  # êµ¬ë§¤ íšŸìˆ˜ ê°€ì¤‘ì¹˜ 20%
    df["êµ¬ë§¤ ë¹ˆë„ ì ìˆ˜"] * 0.2 +  # ìµœê·¼ êµ¬ë§¤ì¼ ê°€ì¤‘ì¹˜ 20%
    df["ê±°ë˜ ë°©ì‹ ì ìˆ˜"] * 0.15 +  # ê±°ë˜ ë°©ì‹ ê°€ì¤‘ì¹˜ 15%
    df["êµ¬ë§¤ ê²½ë¡œ ì ìˆ˜"] * 0.1 +  # êµ¬ë§¤ ê²½ë¡œ ê°€ì¤‘ì¹˜ 10%
    df["ê³ ê° ë“±ê¸‰ ì ìˆ˜"] * 0.1  # ê¸°ì¡´ ê³ ê° ë“±ê¸‰ ì ìˆ˜ í™œìš©
)

# ğŸ”¹ ìµœì¢… ì ìˆ˜ë¥¼ 1~100 ìŠ¤ì¼€ì¼ë¡œ ë³€í™˜ (Min-Max ì •ê·œí™”)
df["ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜"] = df["ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜"] / df["ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜"].max() * 100

# ğŸ”¹ ê²°ê³¼ í™•ì¸
print("\nğŸ“Œ [ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜ ê³„ì‚° ê²°ê³¼]")
print(df[["ê³ ê°ID", "ê¸ˆì•¡ ì ìˆ˜", "êµ¬ë§¤ íšŸìˆ˜ ì ìˆ˜", "êµ¬ë§¤ ë¹ˆë„ ì ìˆ˜", "ê±°ë˜ ë°©ì‹ ì ìˆ˜", "êµ¬ë§¤ ê²½ë¡œ ì ìˆ˜", "ê³ ê° ë“±ê¸‰ ì ìˆ˜", "ê³ ê° ì¶©ì„±ë„ ì§€ìˆ˜"]].head())



```
